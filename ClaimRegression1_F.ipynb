```python
# IQR Outlier Removal for Claim_Amount
Q1 = df['Claim_Amount'].quantile(0.25)
Q3 = df['Claim_Amount'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df_clean = df[(df['Claim_Amount'] >= lower_bound) & (df['Claim_Amount'] <= upper_bound)]

print(f"Original shape: {df.shape}")
print(f"Cleaned shape: {df_clean.shape}")
```

```python
# Replot Boxplots using Cleaned Data
plt.figure(figsize=(18, 6))
for i, col in enumerate(categorical_cols, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(data=df_clean, x=col, y='Claim_Amount', palette='pastel')
    plt.title(f'Claim Amount by {col} (Outliers Removed)')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
x=df[['Age','Annual_Income','Claim_History','Risk_Score','Fraud_Label','Premium_Amount','Policy_Type','Gender']]
x
```

```python
y=df[['Claim_Amount']]
y
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder

le=LabelEncoder()
x['Risk_Score']=le.fit_transform(x['Risk_Score'])
```

```python
x['Risk_Score'].value_counts()
```

```python
oh = OneHotEncoder(sparse_output=False)  
policy_encoded = oh.fit_transform(x[['Policy_Type']])


policy_df = pd.DataFrame(policy_encoded, columns=oh.get_feature_names_out(['Policy_Type'])).astype(int)

x = x.drop(columns=['Policy_Type']).reset_index(drop=True)
x = pd.concat([x, policy_df], axis=1)
```

```python
x
```

```python
gender_encoded=oh.fit_transform(x[['Gender']])
gender_df=pd.DataFrame(gender_encoded,columns=oh.get_feature_names_out(['Gender'])).astype(int)
x=x.drop(columns=['Gender']).reset_index(drop=True)
x=pd.concat([x,gender_df],axis=1)
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import MinMaxScaler
s=MinMaxScaler()
x_scaled=s.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42)
```

```python
x
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Feature scaling
s = MinMaxScaler()
x_scaled = s.fit_transform(x)

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)

# Train model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(x_train, y_train)

# Predict
ypred_test = model.predict(x_test)
ypred_train = model.predict(x_train)

# Evaluate model
print("\nðŸ“Š Evaluation on Test Set:")
print("R2 Score       :", r2_score(y_test, ypred_test))
print("MAE            :", mean_absolute_error(y_test, ypred_test))
print("MSE            :", mean_squared_error(y_test, ypred_test))
print("RMSE           :", np.sqrt(mean_squared_error(y_test, ypred_test)))

print("\nðŸ“Š Evaluation on Train Set:")
print("R2 Score       :", r2_score(y_train, ypred_train))
print("MAE            :", mean_absolute_error(y_train, ypred_train))
print("MSE            :", mean_squared_error(y_train, ypred_train))
print("RMSE           :", np.sqrt(mean_squared_error(y_train, ypred_train)))
```

```python
# Feature importance plot
importances = model.feature_importances_
features = x.columns

plt.figure(figsize=(10, 6))
plt.barh(features, importances, color='skyblue')
plt.title("Feature Importance")
plt.xlabel("Importance")
plt.tight_layout()
plt.show()
```

```python
# Predict for a single input sample
# Example: (Age, Annual_Income, Claim_History, Risk_Score_encoded, Fraud_Label, Premium_Amount,
# Policy_Type_Auto, Policy_Type_Health, Policy_Type_Life, Policy_Type_Property, Gender_Female, Gender_Male)
inputdata = (56,114703,3,2,0,7729,0,0,1,0,1,0)
inputdataasarray = np.asarray(inputdata).reshape(1, -1)
inputdata_scaled = s.transform(inputdataasarray)
prediction = model.predict(inputdata_scaled)
print("\nðŸ”® Predicted Claim Amount:", round(*prediction, 2))

# Optional: Tree-wise prediction range
all_preds = [tree.predict(inputdata_scaled)[0] for tree in model.estimators_]
print("\nPrediction Interval:")
print(f"Min: {min(all_preds):.2f}, Max: {max(all_preds):.2f}, Mean: {np.mean(all_preds):.2f}")
```
