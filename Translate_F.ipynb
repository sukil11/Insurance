```python
#-----------------------------Text preprocessing---------------------------#
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("punkt_tab")
```

```python
stop_words=set(stopwords.words("english"))
lemma=WordNetLemmatizer()
import re
def clean(text):
  if text is None:
    return ''
  text=re.sub(r'https\S+','',text)
  text=re.sub('[^a-zA-Z]'," ",text)
  text=text.lower()
  text=word_tokenize(text)
  text=[lemma.lemmatize(word=w,pos='v')for w in text if w not in stop_words if len(w)>2]
  text=' '.join(text)
  return text
df['cleanedtext']=df['Policy_Text_EN'].apply(clean)
df['cleanedsummary']=df['Summarized_Text_EN'].apply(clean)
```

```python
import pandas as pd
from googletrans import Translator

# Load your dataset
df1 = pd.read_csv("cleaned_insurance_text.csv")
print("✅ CSV file loaded: 'cleaned_insurance_text.csv'")

# Initialize translator
translator = Translator()

# Define helper function with error handling
def translate_text(text, dest_lang):
    try:
        return translator.translate(str(text), dest=dest_lang).text
    except Exception as e:
        print(f"❌ Error translating: {text} -> {dest_lang} | {e}")
        return "Translation Error"

# Translate and create new columns
df1['cleanedtext_tamil'] = df1['cleanedtext'].apply(lambda x: translate_text(x, 'ta'))
df1['cleanedsummary_tamil'] = df1['cleanedsummary'].apply(lambda x: translate_text(x, 'ta'))

df1['cleanedtext_french'] = df1['cleanedtext'].apply(lambda x: translate_text(x, 'fr'))
df1['cleanedsummary_french'] = df1['cleanedsummary'].apply(lambda x: translate_text(x, 'fr'))

# Save to new CSV
df1.to_csv("translated_insurance_text.csv", index=False)
print("✅ Translations saved to 'translated_insurance_text.csv'")
```

```python
from transformers import M2M100ForConditionalGeneration, M2M100Tokenizer
import pandas as pd

# Load model and tokenizer
model_name = "facebook/m2m100_418M"
tokenizer = M2M100Tokenizer.from_pretrained(model_name)
model = M2M100ForConditionalGeneration.from_pretrained(model_name)

# Helper function to translate a sentence
def translate_text(text, src_lang, tgt_lang):
    tokenizer.src_lang = src_lang
    encoded = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    generated_tokens = model.generate(
        **encoded,
        forced_bos_token_id=tokenizer.get_lang_id(tgt_lang)
    )
    return tokenizer.decode(generated_tokens[0], skip_special_tokens=True)

# Load your data
df1 = pd.read_csv("cleaned_insurance_text.csv")

# Translate cleanedtext
df1["cleanedtext_ta"] = df1["cleanedtext"].apply(lambda x: translate_text(str(x), "en", "ta"))
df1["cleanedtext_fr"] = df1["cleanedtext"].apply(lambda x: translate_text(str(x), "en", "fr"))

# Translate cleanedsummary
df1["cleanedsummary_ta"] = df1["cleanedsummary"].apply(lambda x: translate_text(str(x), "en", "ta"))
df1["cleanedsummary_fr"] = df1["cleanedsummary"].apply(lambda x: translate_text(str(x), "en", "fr"))

# Save to CSV
df1.to_csv("translated_insurance_text1.csv", index=False, encoding="utf-8")

print("✅ Translation complete! File saved as translated_insurance_text.csv")
```
